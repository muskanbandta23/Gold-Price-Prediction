Question 28: Find Minumum and Maximum value of gold over years and represent it graphically?

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
high_per_year = df.groupby('year')['High'].max()
low_per_year = df.groupby('year')['Low'].min()
# Concatenating two Series
s4 = pd.concat([high_per_year, low_per_year], axis=1, keys=['High', 'Low'])
print("\nConcatenated Series using concat():\n", s4)
plt.plot(high_per_year.index, high_per_year.values, color='orange')
plt.plot(low_per_year.index, low_per_year.values, color='red')
plt.xlabel('High')
plt.ylabel('Low')
plt.show()
```

 Question 29 : Find Average value of gold according to their open and close price and represent it graphically?

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
open_per_year = df.groupby('year')['Open'].mean().to_frame()
close_per_year = df.groupby('year')['Close'].mean().to_frame()
# Joining two Series
s3 = open_per_year.join(close_per_year, how='outer', lsuffix='open_per_year', rsuffix='close_per_year')
print("Joined Series using join():\n", s3)
s3.plot(y=['Open' , 'Close'])
plt.show()
```

Question 30 . Find variance value of gold according to their open and close price and represent it graphically?

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
open_per_year = df.groupby('year')['Open'].var().to_frame()
close_per_year = df.groupby('year')['Close'].var().to_frame()
# merge
s2 = pd.merge(open_per_year, close_per_year, how='outer', left_index=True, right_index=True, suffixes=('open_per_year', 'close_per_year'))
print("\nMerged Series using merge():\n", s2)
s3.plot(y=['Open' , 'Close'])
plt.show()
```
